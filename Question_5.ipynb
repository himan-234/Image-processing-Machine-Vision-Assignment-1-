```python
image = cv2.imread("rice_salt_pepper_noise.png", cv2.IMREAD_GRAYSCALE)

kernel_open = np.ones((5, 5), np.uint8)
image_opened = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel_open, iterations=2)

kernel_close = np.ones((5, 5), np.uint8)
image_closed = cv2.morphologyEx(image_opened, cv2.MORPH_CLOSE, kernel_close, iterations=10)

plt.subplot(131), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(132), plt.imshow(image_opened, cmap='gray'), plt.title('Opened Image')
plt.subplot(133), plt.imshow(image_closed, cmap='gray'), plt.title('Closed Image')
plt.show()
```

```python
im = cv2.imread('rice_gaussian_noise.png', cv2.IMREAD_GRAYSCALE)

denoised_im = cv2.fastNlMeansDenoising(im, None, h=28, searchWindowSize=10)

_, segmented_image = cv2.threshold(denoised_im, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
closed_image = cv2.morphologyEx(segmented_image, cv2.MORPH_CLOSE, kernel)
opened_image = cv2.morphologyEx(closed_image, cv2.MORPH_OPEN, kernel)

num_labels, labels = cv2.connectedComponents(opened_image)

num_rice_grains = num_labels - 1

print("Number of rice grains:", num_rice_grains)
```
