```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
img = cv2.imread('highlights_and_shadows.jpg', cv2.IMREAD_COLOR)

gamma = 0.5

table = np.array(
    [(i/255.0)**(gamma)*255.0 for i in np.arange(0, 256)]).astype('uint8')

img_gamma = cv2.LUT(img, table)
img_orig = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gamma = cv2.cvtColor(img_gamma, cv2.COLOR_BGR2RGB)

f, axarr = plt.subplots(1, 2)
color = ('b', 'g', 'r')

for i, c in enumerate(color):
    hist_orig = cv2.calcHist([img_orig], [i], None, [256], [0, 256])
    axarr[0].plot(hist_orig, color=c)
    axarr[0].set_title('Original Histogram')
    hist_gamma = cv2.calcHist([img_gamma], [i], None, [256], [0, 256])
    axarr[1].plot(hist_gamma, color=c)
    axarr[1].set_title('Gamma Corrected Histogram')
```
