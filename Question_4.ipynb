```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
```

```python
im = cv.imread('shells.tif', cv.IMREAD_GRAYSCALE)
plt.figure(figsize=[10, 5])
plt.subplot(1, 2, 1)
plt.gca().set_title('Original Histogram')
h = np.zeros(256)
h = [np.sum(im == i) for i in range(256)]
plt.bar(range(256), h)
plt.subplot(1, 2, 2)
plt.gca().set_title('Equalized Histogram')
eh = cv.equalizeHist(im)
plt.hist(eh.flatten(), bins=256, range=[0, 256], density=True)
plt.show()

fig, ax = plt.subplots(1, 2, figsize=(10, 20))
ax[0].imshow(im, cmap="gray")
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(eh, cmap="gray")
ax[1].set_title('Transformed')
ax[1].axis('off')
plt.show()
```
